**1. Data Decription:**

This dataset collected from Netflix of different TV shows and Movies from the year 2008 to 2021.

1.show_id: Gives the information about show id.

2.type: Gives information about 2 different unique values one is TV Show and another is Movie.

3.title: Gives information about the title of Movie or TV Show.
director: Gives information about the director who directed the Movie or TV Show.

4.cast: Gives information about the cast who plays role in Movie or TV Show.
country: Gives information about the Name of country.

5.date_added: Gives information about the tv shows or movie released.
release_year: Gives information about the year when Movie or TV Show was released.

6.rating: Gives information about the Movie or TV Show are in which category (eg like the movies are only for students, or adults, etc).

7.duration: Gives information about the duration of Movie or TV Show.

8.listed_in: Gives information about the genre of Movie or TV Show.

9.description: Gives information about the description of Movie or TV Show.

4.2 What is the Correlation between the features?

```python
netflix_copy.dtypes
```



*   Converting object into datetime formate of "date_added"



```python
netflix_copy["date_added"]=pd.to_datetime(netflix_copy['date_added'],format='mixed')
```

```python
netflix_copy.head(2)
```



*  Now ,we create two features year_added and month_added from date_added column.



```python
netflix_copy['year_added'] = netflix_copy["date_added"].dt.year
netflix_copy['month_added'] = netflix_copy["date_added"].dt.month
netflix_copy.head(2)
```

```python
netflix_copy.dtypes
```

```python
columns_to_plot = ['year_added', 'month_added', 'release_year']
netflix_subset = netflix_copy[columns_to_plot]
correlation_matrix = netflix_subset.corr()
```

```python
plt.subplots(figsize=(5,5))
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

**Obeservation** : Above Heatmap shows correlation between release_year,year_added & month_added.
